```python
import numpy as np
import pandas as pd
df_main = pd.read_csv('new_main.csv')
'''此py将所有附表合并到主表，并且最后得到结果（每个neighbr的分数）
值得注意的是附表合并进来会产生缺少数据的情况，根据不同数据种类，处理方式不同'''
```

```python
df_new = df_main[['borough_code_definition','neighborhood','address','zipcode','police_precinct','commercial_normal', 'commercial_score', 'tax_rate', 'tax_normal',
       'tax_score', 'unit_price', 'price_normal', 'price_score',
       'property_age', 'property_quality_normal', 'property_quality_score']]
```

```python

#df_new1 = df_new.rename(index=str, columns={"property_normal": "property_quality_normal", "property_score": "property_quality_score"})
df_new2 = df_new[['borough_code_definition','neighborhood','address','zipcode','police_precinct', 'commercial_score',
       'tax_score','price_score','property_quality_score']].copy()
```

```python
df_new2['position'] = df_new2['address'] + df_new2['zipcode'].map(lambda x : ','+str(x)+',NYC') 
#增加了position 行用于和附表drive_time和transit_time的合并
df_new2
```

```python
df_Drive_Time = pd.read_csv('Drive_Time.csv',index_col=[0])[['Drive_Time_Score']]
df_new3 = df_new2.merge(df_Drive_Time,how='left',left_on='position',right_index=True).dropna(subset=['Drive_Time_Score']).copy() 
#因为开车时间不应该存在没有数据的情况，所以将没有数据的信息条作为异常数据删除了
#df_new3['Drive_Time_Score'] = df_new3['Drive_Time_Score'].fillna(value=0)
```

```python
df_new3
```

```python
df_Transit_Time = pd.read_csv('Transit_Time.csv',index_col=[0])[['Transit_Time_Score']]
df_new4 = df_new3.merge(df_Transit_Time,how='left',left_on='position',right_index=True)
df_new4['Transit_Time_Score'] = df_new4['Transit_Time_Score'].fillna(value=0) #因为有些地方可能没有公共交通到LIC,所以将缺失数据的信息条填充0
```

```python
df_new4
```

```python
df_edu = pd.read_csv('Education.csv',index_col=[0])[['neighborhood','Edu_Score']]
df_new5 = df_new4.merge(df_edu,how='left',left_on='neighborhood', right_on='neighborhood')
df_new5['Edu_Score'] = df_new5['Edu_Score'].fillna(value=0)
df_new5
```

```python
df_noise = pd.read_csv('Noise.csv',index_col=[0])[['Noise_Score']]
#df_noise
#df_FP.columns = ['facilities_score', 'pm2.5_score']
df_new6 = df_new5.merge(df_noise,how='left',left_on='neighborhood', right_index=True)
df_new6['Noise_Score'] = df_new6['Noise_Score'].fillna(value=df_noise['Noise_Score'].mean()) 
#没有噪音举报数据应该是好事，但由于不确定缺失原因是不是只是因为没有人，所以填充平均分
df_new6
```

```python
df_FP = pd.read_csv('Facilities&PM.csv',index_col=[0])
#df_FP
df_FP.columns = ['facilities_score', 'pm2.5_score']
df_new7 = df_new6.merge(df_FP,how='left',left_on='borough_code_definition', right_index=True)
df_new7
```

```python
df_safe = pd.read_csv('safetyscore.csv',index_col=[0])[['grade']]
df_safe.columns = ['safe_score']
#df_safe
df_new8 = df_new7.merge(df_safe,how='left',left_on='police_precinct', right_index=True)
df_new8
```

```python
df_trafic = pd.read_csv('traffic_data_weighting.csv',index_col=[0])[['score_weighting']]
df_trafic.columns = ['trafic_score']
#df_trafic
df_new9 = df_new8.merge(df_trafic,how='left',left_on='borough_code_definition', right_index=True)
df_new9
```

```python
df_new9.columns
```

```python
df_new9['total_score'] = df_new9['commercial_score'] + df_new9['tax_score'] + df_new9['price_score'] + df_new9['property_quality_score']+\
                         df_new9['Drive_Time_Score'] + df_new9[ 'Transit_Time_Score'] + df_new9['Edu_Score'] + df_new9['Noise_Score']+\
                         df_new9['facilities_score'] + df_new9[ 'pm2.5_score'] + df_new9['safe_score'] + df_new9[ 'trafic_score']   
df_new9
```

```python
df_final = df_new9[['borough_code_definition', 'neighborhood','zipcode', 'total_score']].copy()
df_final['Neighborhood'] = df_final['neighborhood'].map(lambda x : str(x)+', ') + df_final['borough_code_definition']
df_final1 = df_final.groupby(['Neighborhood'])[['total_score']].mean() 
df_final2 = df_final1.sort_values(['total_score'],ascending=False)
#df_pst = df_new9[['borough_code_definition', 'neighborhood','zipcode']].copy()
#df_pst1 = df_pst.drop_duplicates().copy()
# df_result = df_final2.merge(df_pst1,how='left',left_index=True,right_on='neighborhood').copy()
df_final2
#df_pst1
#df_final2.reset_index()
```

```python
df_final_result = df_final2.reset_index()
df_final_result
df_final_result.to_csv('Final_Result.csv')
```
